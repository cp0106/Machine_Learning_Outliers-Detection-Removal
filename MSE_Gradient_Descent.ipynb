Visualization of MSE (Mean Squared Error - Cost function), Gradient Descent and Predicted_Y and Actual_y Comparison

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def g_d(x,y):
    n = len(x)
    m_curr= 0
    b_curr = 0
    l_rate = 0.008

    for i in range(5000):

      y_pred = m_curr*x+b_curr
      mse = (1/n) * sum([i**2 for i in (y-y_pred)])
      md= (-2/n)* sum(x *(y-y_pred))
      bd=(-2/n)* sum((y-y_pred))
      m_curr = m_curr - l_rate*md
      b_curr = b_curr -l_rate*bd
      plt.plot(x,y_pred,color='red', marker = '+', )
      # print(i, m_curr, b_curr, mse)

    print(y_pred)


x = np.array([1,2,3,4,5])
y = np.array([10,14,18,22,26])

g_d(x,y)
```
